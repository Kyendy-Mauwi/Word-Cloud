```python
import random  # Import the random module to generate random words
from wordcloud import WordCloud, STOPWORDS  # Import WordCloud and STOPWORDS from wordcloud
import matplotlib.pyplot as plt  # Import matplotlib for plotting

# Function to generate a random word of a given length
def generate_random_word(length):
    letters = "kevin,kyendy,mauwi,kenya national reit"  # Define the set of letters to choose from
    return ''.join(random.choice(letters) for _ in range(length))  # Generate a random word by randomly selecting letters

comment_words = ''  # Initialize an empty string to accumulate the generated words
stopwords = set(STOPWORDS)  # Get the set of stopwords to exclude from the word cloud

# Number of random words to generate
num_words = 1000  # Define the number of random words to be generated

# Loop to generate the random words
for _ in range(num_words):
    word = generate_random_word(random.randint(3, 10))  # Generate a random word with a random length between 3 and 10 characters
    comment_words += word + " "  # Add the generated word to the comment_words string with a space separating each word

# Generate the word cloud from the accumulated random words
wordcloud = WordCloud(width=800, height=800,  # Set the dimensions of the word cloud
                      background_color='white',  # Set the background color to white
                      stopwords=stopwords,  # Exclude stopwords from the word cloud
                      min_font_size=10).generate(comment_words)  # Generate the word cloud from the comment_words string

# Plot the generated word cloud
plt.figure(figsize=(8, 8), facecolor=None)  # Set up the plotting figure with a square size
plt.imshow(wordcloud)  # Display the word cloud image
plt.axis("off")  # Turn off the axis labels and ticks
plt.tight_layout(pad=0)  # Adjust the layout to remove padding

plt.show()  # Show the plotted word cloud image
```
